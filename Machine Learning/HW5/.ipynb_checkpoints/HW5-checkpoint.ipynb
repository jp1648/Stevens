<h1>Data Plotting and Data Generation</h1>

```python
import matplotlib.pyplot as plt
import numpy as np
noise_scale = 100
number_of_samples = 50
x = 25*(np.random.rand(number_of_samples, 1) - 0.8)
y = 5 * x + 20 * x**2 + 1 * x**3 + noise_scale*np.random.randn(number_of_samples, 1)
plt.plot(x,y,'ro')
plt.grid(True)
```

<h1>Generic Algorithm Implementation</h1>

```python
size_pop = 100
rate_mutation = 0.15
rate_crossover = 0.8
gens = 1000

# Initialize random population of polynomial coefficients
def init_pop(size_pop, chromosome_length):
    return np.random.randn(size_pop, chromosome_length)
```

```python
# MSE function
def fitness_function(individual, x, y):
    y_pred = individual[3] * x**3 + individual[2] * x**2 + individual[1] * x + individual[0]
    return np.mean((y - y_pred) ** 2)

# Select parents based on fitness
def select_parents(population, fitness):
    probabilities = fitness / fitness.sum()
    selected = np.random.choice(np.arange(len(population)), size=len(population), p=probabilities)
    return population[selected]
```

```python
def crossover_operation(parent, parent1, rate_crossover):
    if np.random.rand() < rate_crossover:
        point = np.random.randint(1, len(parent1))
        child = np.concatenate((parent[:point], parent1[point:]))
        child1 = np.concatenate((parent1[:point], parent[point:]))
        return child, child1
    else:
        return parent, parent1

# Apply mutation on individual
def mutate_operation(individual, rate_mutation):
    for i in range(len(individual)):
        if np.random.rand() < rate_mutation:
            individual[i] += np.random.randn()
    return individual
```

```python
# Genetic Algorithm
def genetic_algorithm(x, y, size_pop, rate_mutation, rate_crossover, gens):
    population = init_pop(size_pop, 4)
    best_fits = []

    for generation in range(gens):
        fitness = np.array([1 / (1 + fitness_function(individual, x, y)) for individual in population])
        best_fit = population[np.argmax(fitness)]
        best_fits.append(best_fit)

        # selection
        parents = select_parents(population, fitness)

        # crossover operation and mutation
        next_generation = []
        for i in range(0, size_pop, 2):
            parent = parents[i]
            parent1 = parents[i + 1]
            child, child1 = crossover_operation(parent, parent1, rate_crossover)
            child = mutate_operation(child, rate_mutation)
            child1 = mutate_operation(child1, rate_mutation)
            next_generation.extend([child, child1])

        population = np.array(next_generation)

    return best_fits[-1]
```

```python
best_individual = genetic_algorithm(x, y, size_pop, rate_mutation, rate_crossover, gens)

x_smooth = np.linspace(min(x), max(x), 500)
y_smooth = best_individual[3] * x_smooth**3 + best_individual[2] * x_smooth**2 + best_individual[1] * x_smooth + best_individual[0]

plt.plot(x, y, 'ro')
plt.plot(x_smooth, y_smooth, label='Best Fit', color='blue')
plt.grid(True)
plt.legend()
plt.show()
```

```python
noise_scale = 100
number_of_samples = 50
x = 25 * (np.random.rand(number_of_samples, 1) - 0.8)
y = 5 * x + 20 * x**2 + 1 * x**3 + noise_scale * np.random.randn(number_of_samples, 1)

def plot_noisy_data_and_polynomial(x, y, degree, ax):
    coeffs = np.polyfit(x.flatten(), y.flatten(), degree)  #use polyfit to generate coeffecients for each x var in the polynomial.
    p = np.poly1d(coeffs) #create polynomial function 
    
    x_plot = np.linspace(min(x), max(x), 100) ##generate 100 evenly spaced values between min and max x.
    y_plot = p(x_plot) #using polynomial func, plug in x values to get ys. 
    
    ax.plot(x, y, 'ro', label='Noisy data')
    ax.plot(x_plot, y_plot, label=f'Polynomial degree {degree}')
    ax.legend()

def calculate_mse(x, y, degree):
    coeffs = np.polyfit(x.flatten(), y.flatten(), degree)  #use polyfit to generate coeffecients for each x var in the polynomial.
    p = np.poly1d(coeffs) #create polynomial function 
    y_pred = p(x) #y values from coeff
    mse = np.mean((y - y_pred)**2) #mse calc
    return mse

#Plot noisy data and polynomial
fig, ax = plt.subplots()
degree = 8
plot_noisy_data_and_polynomial(x, y, degree, ax)  
plt.title(f'Noisy data and polynomial {degree}')
plt.show()

# Step 2: Plot MSE vs polynomial degree
degrees = range(1, 9) 
mses = [calculate_mse(x, y, degree) for degree in degrees] #generate MSE for each degree

fig, ax = plt.subplots()
ax.plot(degrees, mses, 'mo-')
plt.title('MSE vs Polynomial Degree')
plt.xlabel('Polynomial Degree')
plt.ylabel('MSE')
plt.show()

best_degree = degrees[np.argmin(mses)]
print(f'Best degree: {best_degree}')
```

```python
fig, ax = plt.subplots()
plot_noisy_data_and_polynomial(x, y, degree=8, ax=ax)  # Polynomial fit
ax.plot(x_smooth, y_smooth, label='Genetic Algorithm Fit', color='blue')  # Genetic Algorithm fit
ax.legend()
plt.title('Noisy data and Polynomial & Genetic Algorithm Fits')
plt.grid(True)
plt.show()
```

np.polyfit tends to be more accurate, more computationally efficient, and predictable for standard polynomial regression tasks. This is especially true when the dataset is predictable, and well-mannered and the relationship between variables is actually polynomial. You can see that the polynomial with degree 8 is more steep and both ends through the maximum and minimums, which can maybe hint towards an overfitting. This tries to fit more towards the noise and outliers. On the other hand, genetic algorithms offer more flexibility for larger data sets and can  discover better fits for more complex data as well as noisy data patterns. This although is at the cost of higher computational costs and the risk of not converging to global optimum. Here, you can see it is doing a better job of generalizing the fit, and is staying away from actually overfitting the model.
